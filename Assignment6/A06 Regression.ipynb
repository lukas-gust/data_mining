# A06 Regression

```python
import numpy as np
```

```python
def read(fname):
    vals = []
    with open(fname, 'rb') as fd:
        for line in fd:
            if not line.strip():
                continue
            s = list(map(float, line.strip().split()))
            vals.append(s)
            
    return np.array(vals)
```

```python
X = read('./X.dat')
print(X.shape)

Y = read('./Y.dat')
print(Y.shape)
```

## Q1 A

```python
A, r , _, __ = np.linalg.lstsq(X, Y)
```

```python
print('Coefficients: \n' + str(A))
```

```python
print('Squared Residual: ' + str(r[0]))
print('Residual: ' + str(r[0]**.5))
```

```python
def ridge_regress(x, y, s):
    A = np.linalg.inv(x.T.dot(x) + (s**2)*np.eye(15)).dot(x.T).dot(y)/s
    A *= s
    r = np.linalg.norm(y - x.dot(A))
    return A, r
```

```python
ss = [1,5,10,15,20,25,30]

for s in ss:
    A, r = ridge_regress(X, Y, s)
    
    print('s: ' + str(s))
    print('Coeffs: \n' + str(A))
    print('Residual: ' + str(r))
    print()
```

## Q1 B

```python
X1 = X[0:66]
X1_C = X[66:]

Y1 = Y[0:66]
Y1_C = Y[66:]

X2 = X[34:100]
X2_C = X[0:34]

Y2 = Y[34:100]
Y2_C = Y[0:34]

X3 = np.vstack([X[0:33], X[67:100]])
X3_C = X[33:67]

Y3 = np.vstack([Y[0:33], Y[67:100]])
Y3_C = Y[33:67]
```

```python
A1, r1, _, _ = np.linalg.lstsq(X1, Y1)
print('Coeffs: \n' + str(A1))
print('Residual: ' + str(r1[0]**.5))
cr1 = np.linalg.norm(Y1_C - X1_C.dot(A1))
print('Cross-Val Residual: ' + str(cr1))
```

```python
A2, r2, _, _ = np.linalg.lstsq(X2, Y2)
print('Coeffs: \n' + str(A2))
print('Residual: ' + str(r2[0]**.5))
cr2 = np.linalg.norm(Y2_C - X2_C.dot(A2))
print('Cross-Val Residual: ' + str(cr2)) 
```

```python
A3, r3, _, _ = np.linalg.lstsq(X3, Y3)
print('Coeffs: \n' + str(A3))
print('Residual: ' + str(r3[0]**.5))
cr3 = np.linalg.norm(Y3_C - X3_C.dot(A3))
print('Cross-Val Residual: ' + str(cr3)) 
```

```python
cr_avg = (cr1+cr2+cr3)/3
cr_avg
```

### Ridge Regression 

```python
crr1 = []
for s in ss:
    A1, r1 = ridge_regress(X1, Y1, s)
    print('s: ' + str(s))
    print('Coeffs: \n' + str(A1))
    print('Residual: ' + str(r1))
    cr1 = np.linalg.norm(Y1_C - X1_C.dot(A1))
    crr1.append(cr1)
    print('Cross-Val Residual: ' + str(cr1))
    print()
```

```python
crr2 = []
for s in ss:
    A2, r2 = ridge_regress(X2, Y2, s)
    print('s: ' + str(s))
    print('Coeffs: \n' + str(A2))
    print('Residual: ' + str(r2))
    cr2 = np.linalg.norm(Y2_C - X2_C.dot(A2))
    crr2.append(cr2)
    print('Cross-Val Residual: ' + str(cr2))
    print()
```

```python
crr3 = []
for s in ss:
    A3, r3 = ridge_regress(X3, Y3, s)
    print('s: ' + str(s))
    print('Coeffs: \n' + str(A3))
    print('Residual: ' + str(r3))
    cr3 = np.linalg.norm(Y3_C - X3_C.dot(A3))
    crr3.append(cr3)
    print('Cross-Val Residual: ' + str(cr3))
    print()
```

```python
ridge_avg = np.vstack([crr1, crr2, crr3]).mean(axis=0)
print(ridge_avg.min())
print(ss[ridge_avg.argmin()])
```

The approach with the lowest average cross-validation residual is ridge regression with $s=20$.

```python
W = read('./W.dat')
print(W.shape)

M = read('./M.dat')
print(M.shape)
```

```python
r = W
s = np.zeros(100)
for _ in range(1000):
    j = np.argmax(r.T.dot(M))
    s[j] = 1
    r = r - np.reshape(M[:,j], (80,1))
    
    print('Epoch: ' + str(_))
    print('Entry: ' + str(j))
    print(np.linalg.norm(r))
    print()
    if not np.count_nonzero(r):
        break
    
print(s)
```
