```python
def mat_pow(P, t, recursive=False):
    if recursive:
        if t == 0:
            return P
        else:
            Pi = mat_pow(P, t//2, recursive=recursive)
            print(np.all(Pi >= 0))
            return Pi @ Pi
        
    else:
        Pi = P
        for i in range(t):
            print(np.all(Pi >= 0))
            Pi = Pi @ P
            
        return Pi
```

```python
def state_prop(P, t, q0):
    q_star = q0
    for i in range(t):
        q_star = P @ q_star
        
    return q_star
```

```python
state_prop(P, t, q0)
```

```python
def rand_walk(P, t, t0, q0):
    # burn in
    q_star = q0.astype(int)
    for i in range(t0):
        W = np.sum(P[:,np.flatnonzero(q_star)])
        w = np.cumsum(P[:,np.flatnonzero(q_star)]/W)
        p = rand.uniform(0,1)
        k = np.argmax(w>p)
        
        q_star = np.zeros(q_star.shape, dtype=int)
        q_star[k] = 1
        
    # walk
    counts = np.zeros(q_star.shape)
    for j in range(t):
        W = np.sum(P[:,np.flatnonzero(q_star)])
        w = np.cumsum(P[:,np.flatnonzero(q_star)]/W)
        p = rand.uniform(0,1)
        k = np.argmax(w>p)
        counts[k] += 1
        
        q_star = np.zeros(q_star.shape, dtype=int)
        q_star[k] = 1
        
        
    return counts/np.linalg.norm(counts, ord=1)
        
```

```python
t0 = 100
rand_walk(P, t, t0, q0)
```
